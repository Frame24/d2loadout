```python
import pandas as pd

CSV_PATH = "processed_heroes.csv"
df = pd.read_csv(CSV_PATH)
df
```

```python
df.info()
df.describe(include="all")
```

```python
# Фильтр по позиции (поменяй role на pos 1 .. pos 5)
role = "pos 1"
by_role = df[df["Role"] == role].copy()
by_role
```

```python
# Топ по матчам / WR / D2PT для выбранной позиции (сортируй по нужной колонке)
cols = ["Hero", "Facet", "facet_number", "Matches", "WR", "D2PT Rating", "Role"]
show = [c for c in cols if c in by_role.columns]
by_role[show].sort_values("Matches", ascending=False).head(20)
```

```python
# Топ по WR (минимум матчей — поменяй min_matches если нужно)
min_matches = 200
subset = by_role[by_role["Matches"] >= min_matches]
subset[show].sort_values("WR", ascending=False).head(20)
```

```python
# Сводка по позициям: записей и матчей
df.groupby("Role").agg(
    записей=("Hero", "count"),
    матчей_сумма=("Matches", "sum"),
    матчей_медиана=("Matches", "median"),
).round(0)
```

```python
# Процентили Matches (как в config_processor: 60-й и 75-й)
if "Matches" in df.columns:
    m = df["Matches"].dropna()
    print("60-й процентиль (basic):", int(m.quantile(0.60)))
    print("75-й процентиль (extended):", int(m.quantile(0.75)))
```
