```python
import os
import subprocess
import sys
import ast
import pkgutil

def extract_imports(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        tree = ast.parse(file.read(), filename=file_path)

    imports = set()
    for node in ast.walk(tree):
        if isinstance(node, ast.Import):
            for alias in node.names:
                imports.add(alias.name.split('.')[0])
        elif isinstance(node, ast.ImportFrom):
            imports.add(node.module.split('.')[0])

    return imports

def install_packages(packages):
    # Получаем список встроенных модулей
    std_libs = {module.name for module in pkgutil.iter_modules()}
    
    for package in packages:
        if package in std_libs:
            print(f"'{package}' is a standard library module, skipping installation.")
            continue
        try:
            subprocess.check_call([sys.executable, "-m", "pip", "install", package])
        except subprocess.CalledProcessError as e:
            print(f"Failed to install package {package}: {e}")

if __name__ == "__main__":
    # Укажите путь к вашему Python файлу
    file_path = "./test.py"
    packages = extract_imports(file_path)
    install_packages(packages)
```

```python
def get_build(driver):
    rows = driver.find_element(By.CSS_SELECTOR, '.grid.grid-cols-3.p-4.gap-1.rounded-md.bg-d-gray-5.border.border-solid.border-d-gray-8')\
    .find_elements(By.XPATH, "./child::*")
    rows
    talents = rows[0]\
        .find_elements(By.CSS_SELECTOR, ".abs.svelte-yv4bey")
    talents_classtext = [talent.get_attribute("class") for talent in talents]
    build_talents = [re.findall("(\d+)\s(rt|lt)" , talents_res) for talents_res in talents_classtext]
    build_talents = [t[0] for t in build_talents if t]
    build_talents = [int(x[1] == "rt") for x in sorted(build_talents)]
    build_talents
    abilities = rows[0]\
        .find_element(By.CSS_SELECTOR, '.flex.flex-wrap.gap-1.items-center').find_elements(By.XPATH, "./child::img")
    build_abilities = [a.get_attribute("title") for a in abilities]
    build_abilities
    starting = rows[1]\
        .find_elements(By.CSS_SELECTOR, '.flex.flex-col.gap-2.bg-d2pt-gray-1.p-2.rounded-md.border-one')
    starting = [div.find_elements(By.XPATH, "./child::div")[0] for div in starting]
    build_starting = [a.get_attribute("title") for a in starting]
    build_starting
    common_late = rows[2]\
        .find_elements(By.CSS_SELECTOR, '.text-xs.text-white.font-medium.text-right.rounded-md.text-shadow')
    build_common_late = [a.get_attribute("title") for a in common_late]
    build_common_late
    rows_2 = driver.find_elements(By.CSS_SELECTOR, '.grid.grid-cols-3.gap-1.rounded-md.bg-d-gray-5.border.border-solid.border-d-gray-8')[1]\
    .find_elements(By.XPATH, "./child::*")
    rows_2
    early = rows_2[0]\
        .find_elements(By.CSS_SELECTOR, '.flex.flex-col.gap-2.bg-d2pt-gray-1.p-2.rounded-md.border-one')
    early = [div.find_elements(By.XPATH, "./child::div")[0] for div in early]
    build_early = [a.get_attribute("title") for a in early]
    build_early
    mid = rows_2[1]\
        .find_elements(By.CSS_SELECTOR, '.flex.flex-col.gap-2.bg-d2pt-gray-1.p-2.rounded-md.border-one')
    mid = [div.find_elements(By.XPATH, "./child::div")[0] for div in mid]
    build_mid = [a.get_attribute("title") for a in mid]
    build_mid
    late = rows_2[2]\
        .find_elements(By.CSS_SELECTOR, '.flex.flex-col.gap-2.bg-d2pt-gray-1.p-2.rounded-md.border-one')
    late = [div.find_elements(By.XPATH, "./child::div")[0] for div in late]
    build_late = [a.get_attribute("title") for a in late]
    build_late
    core = driver.find_element(By.CSS_SELECTOR, '.flex.mb-2.flex-col.p-4.gap-2.rounded-md.bg-d-gray-5.border.border-solid.border-d-gray-8.overflow-x-auto')\
    .find_elements(By.CSS_SELECTOR, '.flex.gap-1.flex-col.bg-d2pt-gray-1.p-1.rounded-md.border-one.text-xs.text-white.font-medium')
    build_core = [c.find_elements(By.XPATH, "./child::div")[1].get_attribute("title") for c in core]
    build_core
    return [build_talents, build_abilities, build_starting, build_common_late, build_early, build_mid, build_late, build_core]
```

```python
# Функция для создания нового экземпляра драйвера
def make_driver():
    return webdriver.Chrome(options=make_options())

# Функция для ожидания изменения текста элемента
def text_to_change(driver, child_element, initial_text):
    def _predicate(driver):
        parent_element = driver.execute_script("return arguments[0].parentNode;", child_element)
        return parent_element.text != initial_text
    WebDriverWait(driver, 30).until(_predicate)

# Функция для парсинга данных
def parse_site(link, driver, arr_df):
    try:
        driver.get(link)
        hero_name = link.split("/")[-1].replace("%20", " ")
        WebDriverWait(driver, 30).until(
            EC.presence_of_element_located(
                (By.CSS_SELECTOR, ".flex.flex-col.gap-2.sticky.top-1")
            )
        )
        positions = driver.find_element(By.CSS_SELECTOR, ".flex.flex-col.gap-2.sticky.top-1").find_elements(By.XPATH, "./child::*")[1:]
        for position in positions:
            
            # Ваш код для поиска элемента
            child_element = driver.find_elements(By.CSS_SELECTOR, ".flex.flex-col.gap-2.p-2.bg-d-gray-5.border.border-solid.border-d-gray-8.rounded-md")[-1]
            parent_element = driver.execute_script("return arguments[0].parentNode;", child_element)

            # Текущий текст элемента
            initial_text = parent_element.text

            # Запускаем ожидание в отдельном потоке
            wait_thread = Thread(target=text_to_change, args=(driver, child_element, initial_text))
            wait_thread.start()

            # Выполняем клик
            position.click()

            # Ожидаем завершения потока
            wait_thread.join()
            
            position_rating = driver.find_element(By.CSS_SELECTOR, '.flex.justify-center.gap-1.text-yellow.text-xs.items-center.d-box-1.p-2').text.split(" ")[2].split("\n")[0]
            position_matches = re.findall(r"\d+ matches", position.text)[0].split(" ")[0]
            position_win_rate = re.findall(r"\d+.\d+%", position.text)[0].split("%")[0]
            position_text = position.text.split("\n")[0]
            facets = driver.find_elements(By.CSS_SELECTOR, '.cursor-pointer.flex.flex-col.rounded-md.relative')
            for facet_number, facet in enumerate(facets, start=1):
                facet.click()
                facet_matches = re.findall(r"\d+ matches", facet.text)[0].split(" ")[0]
                facet_win_rate = re.findall(r"\d+.\d+%", facet.text)[1].split("%")[0]
                facet_text = facet.text.split("\n")[-1]
                builds = driver.find_elements(By.CSS_SELECTOR, '.rounded-md.svelte-15r7zt1')
                for build in builds:
                    build.click()
                    build_matches = re.findall(r"\d+ matches", build.text)[0].split(" ")[0]
                    build_win_rate = re.findall(r"\d+% win rate", build.text)[0].split("%")[0]
                    arr_df.append([hero_name, position_text, position_matches, position_win_rate,position_rating, facet_text, facet_number, facet_matches, facet_win_rate, build_matches, build_win_rate] + get_build(driver))
    except Exception as e:
        logging.error(f"Error processing link {link}: {e}")

# Функция для обработки задач
def worker(driver, task_queue, arr_df, progress_bar):
    while not task_queue.empty():
        link = task_queue.get()
        try:
            parse_site(link, driver, arr_df)
        except:
            print(f"An error occured while parsing {link}")
        finally:
            task_queue.task_done()
            progress_bar.update(1)

# Основная функция для запуска парсинга
def main(links):
    # Количество драйверов
    drivers_num = 1

    # Создание экземпляров драйверов
    drivers = [make_driver() for _ in range(drivers_num)]

    # Создание очереди задач и добавление ссылок
    task_queue = Queue()
    for link in links:
        task_queue.put(link)

    # Массив для хранения результатов
    arr_df = []

    # Создание прогресс-бара
    with tqdm(total=len(links), desc="Processing links", file=sys.stdout, ascii=True) as progress_bar:
        # Создание и запуск потоков
        threads = []
        for i in range(drivers_num):
            thread = Thread(target=worker, args=(drivers[i], task_queue, arr_df, progress_bar))
            thread.start()
            threads.append(thread)

        # Ожидание завершения всех потоков
        for thread in threads:
            thread.join()

    # Закрытие драйверов
    for driver in drivers:
        # Закрытие драйвера
        driver.close()
        driver.quit()
        
    # df = pd.DataFrame(arr_df)
    # Преобразование результатов в DataFrame
    df = pd.DataFrame(arr_df, columns=["hero_name", "position_text", "position_matches", "position_win_rate", "position_rating", "facet_text", "facet_number", "facet_matches", "facet_win_rate", "build_matches", "build_win_rate", "talents", "abilities", "starting", "common_late", "early", "mid", "late", "core"])

    # Сохранение DataFrame в файл
    # df.to_csv('results.csv', index=False)

    return df
```

```python
def get_hero_links(driver, base_url):
    driver.get(base_url)

    # Явное ожидание, чтобы убедиться, что вся страница загружена
    WebDriverWait(driver, 30).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "a[href*='/hero/']"))
    )

    # Поиск всех ссылок на героев
    hero_links = []
    links = driver.find_elements(By.CSS_SELECTOR, "a[href*='/hero/']")
    for link in links:
        href = link.get_attribute('href')
        # Исключение конкретной ссылки
        if href != "https://dota2protracker.com/hero/" and href not in hero_links:
            hero_links.append(href)

    return hero_links

# Основной URL сайта
base_url = "https://dota2protracker.com"

# Создание экземпляра драйвера
driver = make_driver()

# Получение ссылок на всех героев
hero_links = get_hero_links(driver, base_url)

# Закрытие драйвера
driver.close()
driver.quit()

# hero_links
```

```python
def get_by_rating(df: pd.DataFrame, position: int):
    return (
        df[(df["pos"] == position) & (df["position_matches"] > 100)]
        .sort_values(["position_rating"], ascending=False)["hero_id"]
        .values
        .tolist()
    )
```

```python
def get_popular(df: pd.DataFrame, position: int):
    return (
        df[(df["pos"] == position) & (df["position_matches"] > 100)]
        .sort_values(["position_matches"], ascending=False)["hero_id"]
        .values
        .tolist()
    )
```

```python
def get_wr(df: pd.DataFrame, position: int):
    return (
        df[(df["pos"] == position) & (df["position_matches"] > 100) & (df["position_win_rate"] > 50)]
        .sort_values(["position_win_rate"], ascending=False)["hero_id"]
        .values.tolist()
    )


def get_others_wr(df: pd.DataFrame):
    df_temp = df.groupby("hero_name").max()
    return [
        df_temp[(df_temp["position_matches"] > 100) & (df_temp["position_win_rate"] < 50) & (df_temp["position_win_rate"] > 48)]
        .sort_values(["position_win_rate"], ascending=False)["hero_id"]
        .values.tolist(),
        
        df_temp[(df_temp["position_matches"] > 100) & (df_temp["position_win_rate"] < 48) & (df_temp["position_win_rate"] > 45)]
        .sort_values(["position_win_rate"], ascending=False)["hero_id"]
        .values.tolist(),
        
        df_temp[(df_temp["position_matches"] > 100) & (df_temp["position_win_rate"] < 45)]
        .sort_values(["position_win_rate"], ascending=False)["hero_id"]
        .values.tolist(),
    ]
```
